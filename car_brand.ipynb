```python
from tensorflow.keras.layers import Input, Dense, Flatten, Lambda
from tensorflow.keras.models import Model,load_model
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
Imsize=[224,224]
Path_train='C:/Users/reena/Desktop/python learning files/Car Brand-Transfer Learning/Datasets/Train'
Path_test='C:/Users/reena/Desktop/python learning files/Car Brand-Transfer Learning/Datasets/Test'
```

```python
RN=ResNet50(input_shape=Imsize+[3],weights='imagenet',include_top=False)
```

```python
RN.summary()
```

```python
for layer in RN.layers:
    layer.trainable=False
```

```python
Dataset_folders=glob('C:/Users/reena/Desktop/python learning files/Car Brand-Transfer Learning/Datasets/train/*')
Dataset_folders
```

```python
Flat_RNoutput=Flatten()(RN.output)
```

```python
Prediction=Dense(len(Dataset_folders),activation='softmax')(Flat_RNoutput)
```

```python
My_model=Model(inputs=RN.input,outputs=Prediction)
```

```python
My_model.summary()
```

```python
My_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
Augment_train=ImageDataGenerator(rescale=1.0/255,
                                 shear_range=0.2,
                                 zoom_range=0.2,
                                 horizontal_flip=True,
                                 vertical_flip=False
                                )
```

```python
Scaled_Test=ImageDataGenerator(rescale=1.0/255)#just rescale test data nothing else
```

```python
train_data=Augment_train.flow_from_directory(Path_train,target_size=(224,224),batch_size=32,class_mode='categorical')
```

```python
test_data=Augment_train.flow_from_directory(Path_test,target_size=(224,224),batch_size=32,class_mode='categorical')
```

```python
fit_model=My_model.fit(train_data,validation_data=test_data,epochs=100,steps_per_epoch=len(train_data),validation_steps=len(test_data))
```

```python
fit_model.history
```

```python
#Loss plot
plt.plot(fit_model.history['loss'],label='training_loss')
plt.plot(fit_model.history['val_loss'],label='validation_loss')
plt.legend()
plt.show()
plt.savefig('loss')
```

```python
#accuracy plot
plt.plot(fit_model.history['accuracy'],label='training_accuracy')
plt.plot(fit_model.history['val_accuracy'],label='validation_accuracy')
plt.legend()
plt.show()
plt.savefig('accuracy')
```

```python
My_model.save('RNcarbrand.h5')
```

```python
Img_prediction=My_model.predict(test_data)
```

```python
Prediction=np.argmax(Img_prediction,axis=1)
Prediction
```
